```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
rv = sts.gamma(2)
sample = rv.rvs(100)
print(sample)
```

```python
x = np.linspace(0,10,1000)

cdf = rv.cdf(x) # функция может принимать и вектор (x)

plt.plot(x, cdf)
plt.hist(cdf,bins=10,normed=True)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()

```
