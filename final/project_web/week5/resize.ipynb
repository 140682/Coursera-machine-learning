```python
# Вырезаем ИД пользователя из имени файла
def get_user_id(filename):
    start_pos=(filename.find('\\user'))+5
    end_pos=(filename.find('.csv'))
    return int(filename[start_pos:end_pos])
```

```python
def resize_df(df,window_size):
    res=pd.DataFrame()
    # Добавляем колонки
    for col in range(1,11):
        shift_index=col-1
        site_col_name='site'+str(col)
        res[site_col_name]=df.site[shift_index:]
        res[site_col_name]=res[site_col_name].shift(-shift_index)
        timestamp_col_name='time'+str(col)
        res[timestamp_col_name]=df.timestamp[shift_index:]
        res[timestamp_col_name]=res[timestamp_col_name].shift(-shift_index)
    # Удаляем лишние строки
    ind_tosave=[x for x in range(0,len(df),window_size)]
    return res.iloc[ind_tosave,:]
```

```python
# Функция возвращает подготовленные данные и словарь сайтов
def prepare_train_set(path_to_csv_files, window_size=10):
     # Последовательно читаем все файлы сессий в каталоге и склеиваем их в один DataFrame
    sessions_df=pd.DataFrame()
    filelist=glob.glob(path_to_csv_files+'*.csv')
    for filename in tqdm(filelist):
        df=process_user(filename,window_size)
        sessions_df=sessions_df.append(df)
    print('Формируем словарь')
    freq_df=pd.concat([sessions_df.site1,sessions_df.site2],ignore_index=True,axis=0)
    return sessions_df,freq_df
```
